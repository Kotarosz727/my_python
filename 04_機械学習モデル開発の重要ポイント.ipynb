```python
# 多値ラベルの数値化
# One-hotエンコーディング
w = pd.get_dummies(df9['男女子供'], prefix=('男女子供'))
w.head(20)
```

```python
# One-hotエンコーディングを列に追加、元のデータを削除
def encoding(df, column):
  df_encoded = pd.get_dummies(df[column], prefix=column)
  df_dropped = df.drop([column], axis=1)
  new_def = pd.concat([df_dropped, df_encoded], axis=1)

  return new_def  
```

```python
df10 = encoding(df9, '男女子供')
df11 = encoding(df10, '乗船港コード')
df12 = encoding(df11, 'デッキ')

df12.head(10)
```

```python
# 混同行列作成コード
from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
def custom_confusion_matrix(matrix, columns):
  n = len(columns)

  act = ['正解データ'] * n
  pred = ['予測結果'] * n

  cm = pd.DataFrame(matrix, columns=[pred, columns], index=[act, columns])
  return cm
```

```python
cm = custom_confusion_matrix(matrix, ['良性', '悪性'])
cm
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

x = cancer.data
y = cancer.target

test_size = 0.1

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=1)

print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
# 複数のアルゴリズムで精度を比較
def prepare_algorithms(random_seed):
  # 線形回帰
  from sklearn.linear_model import LogisticRegression
  algorithm1 = LogisticRegression(random_state=random_seed)

  # サポートベクターマシン（カーネル）
  from sklearn.svm import SVC
  algorithm2 = SVC(kernel='rbf', random_state=random_seed)

  # 決定木
  from sklearn.tree import DecisionTreeClassifier
  algorithm3 = DecisionTreeClassifier(random_state=random_seed)

  # ランダムフォレスト
  from sklearn.ensemble import RandomForestClassifier
  algorithm4 = RandomForestClassifier(random_state=1)

  # XGBoost
  from xgboost import XGBClassifier
  algorithm5 = XGBClassifier(random_state=random_seed)

  algorithms = [algorithm1, algorithm2, algorithm3, algorithm4, algorithm5]
  return algorithms
```

```python
# 複数アルゴリズムで精度比較
for algo in prepare_algorithms(9):
  algo.fit(x_train, y_train)
  score = algo.score(x_test, y_test)
  name = algo.__class__.__name__
  print(f'score: {score:.4f}  {name}')
```

```python
# ハイパーパラメータの最適化
```

```python
from sklearn.svm import SVC
algorithm = SVC(kernel='rbf', random_state=1)
print(algorithm)
```

```python
# gammaの最適化
gammas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]

for gamma in gammas:
  algo = SVC(kernel='rbf', gamma = gamma,random_state=1)
  algo.fit(x_train, y_train)
  score = algo.score(x_test, y_test)
  print(f'score: {score:.4f}  gamma: {gamma}')

gamma = 0.001
```

```python
# Cの最適化
Cs = [1, 10, 100, 1000, 10000]

for C in Cs:
  algo = SVC(kernel='rbf', gamma = gamma,random_state=1, C=C)
  algo.fit(x_train, y_train)
  score = algo.score(x_test, y_test)
  print(f'score: {score:.4f}  C: {C}')
```

```python
# 交差検定法(cross validation)
```

```python
# 複数のアルゴリズムで精度を比較
def prepare_algorithms2(random_seed):
  # 線形回帰
  from sklearn.linear_model import LogisticRegression
  algorithm1 = LogisticRegression(random_state=random_seed)

  # サポートベクターマシン（カーネル）
  from sklearn.svm import SVC
  algorithm2 = SVC(kernel='rbf', random_state=random_seed, gamma = 0.001, C=1)

  # 決定木
  from sklearn.tree import DecisionTreeClassifier
  algorithm3 = DecisionTreeClassifier(random_state=random_seed)

  # ランダムフォレスト
  from sklearn.ensemble import RandomForestClassifier
  algorithm4 = RandomForestClassifier(random_state=1)

  # XGBoost
  from xgboost import XGBClassifier
  algorithm5 = XGBClassifier(random_state=random_seed)

  algorithms = [algorithm1, algorithm2, algorithm3, algorithm4, algorithm5]
  return algorithms
```

```python
from sklearn.model_selection import StratifiedKFold
stratifiedKFold = StratifiedKFold(n_splits=3)

# cross_validation
from sklearn.model_selection import cross_val_score
for algo in prepare_algorithms2(1):
  scores = cross_val_score(algo, x_train, y_train, cv=stratifiedKFold)
  score = scores.mean()
  name = algo.__class__.__name__
  print(f'平均スコア: {score:.4f} 個別スコア:{scores} {name}')
```
