**データ前処理**

```python
# 不要な項目の削除

# 等室と同じため
df1 = df_titanic.drop('等室名', axis=1)
# 乗船港コードと同じため
df2 = df1.drop('乗船港', axis=1)
# 生存と同じため
df3 = df2.drop('生存可否', axis=1)

df3.head()
```

```python
# 欠損値の対応
df3.isnull().sum()
```

```python
import numpy as np
# 欠損値がある行を削除（二件だけのため削除で対応）
df4 = df3.dropna(subset=['乗船港コード'])

# 平均値で代用
age_ave = df4['年齢'].mean()
df5 = df4.fillna({'年齢': age_ave})

# ダミーデータ'N'を使用
df6 = df5.replace({'デッキ': {np.nan: 'N'}})
```

```python
df6.isnull().sum()
```

```python
df6.head()
```

```python
# 2値ラベルの数値化
```

```python
male_female = {'male':1 , 'female':0}

df7 = df6.copy()
df7['性別'] = df7['性別'].map(male_female)

df7.head()
```

```python
true_false = {True: 1, False:0}

df8 = df7.copy()
df8['成人男性'] = df8['成人男性'].map(true_false)

df9 = df8.copy()
df9['独身'] = df9['独身'].map(true_false)

df9.head()
```

```python
# 多値ラベルの数値化
# One-hotエンコーディング
w = pd.get_dummies(df9['男女子供'], prefix=('男女子供'))
w.head(20)
```

```python
# One-hotエンコーディングを列に追加、元のデータを削除
def encoding(df, column):
  df_encoded = pd.get_dummies(df[column], prefix=column)
  df_dropped = df.drop([column], axis=1)
  new_def = pd.concat([df_dropped, df_encoded], axis=1)

  return new_def  
```

```python
df10 = encoding(df9, '男女子供')
df11 = encoding(df10, '乗船港コード')
df12 = encoding(df11, 'デッキ')

df12.head(10)
```

**評価**

```python
# 混同行列(confusion matrix)
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

x = cancer.data
# 良性0, 悪性１に値を変える 
y = 1 - cancer.target

# 入力データを２次元に絞り込み
x2 = x[:, :2]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x2, y, train_size=0.7, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
algo = LogisticRegression(random_state=1)

algo.fit(x_train, y_train)

y_pred = algo.predict(x_test)

score = algo.score(x_test, y_test)

print(f'score: {score:.4f}')
# score
```

```python
# 混同行列作成コード
from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
def custom_confusion_matrix(matrix, columns):
  n = len(columns)

  act = ['正解データ'] * n
  pred = ['予測結果'] * n

  cm = pd.DataFrame(matrix, columns=[pred, columns], index=[act, columns])
  return cm
```

```python
cm = custom_confusion_matrix(matrix, ['良性', '悪性'])
cm
```

```python
# 適合率、再現率、F値の計算コード
from sklearn.metrics import precision_recall_fscore_support

precision, recall, fscore, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')

print(f'適合率: {precision:.4f}')
print(f'再現率: {recall:.4f}')
print(f'F値: {fscore:.4f}')
```
